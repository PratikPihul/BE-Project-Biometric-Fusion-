```python
import tensorflow as tf
from tensorflow.keras import models
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

model = models.load_model('model_opt.h5')

CATEGORIES = ["Hitesh", "Labhesh", "Tarun", "Varkha"]

img = cv2.imread('./predict/Varkha/Varkha_40_contrast.png', cv2.IMREAD_GRAYSCALE)
plt.imshow(mpimg.imread('./predict/Varkha/Varkha_40_contrast.png'))
img = cv2.resize(img, (60, 100))
img = np.array(img).reshape(-1, 60, 100, 1)
output = model.predict([img])
output = tf.math.argmax(output, axis=1).numpy()
print(CATEGORIES[output[0]])
```
